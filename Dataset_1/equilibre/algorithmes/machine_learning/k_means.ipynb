```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.neighbors import LocalOutlierFactor

# Chargement du dataset
df = pd.read_csv('../final_cleaned.csv')

# Sparation des features (X) et de la variable cible (y)
X = df.drop('isFraud', axis=1)

# Dtection des bruits avec LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # Vous pouvez ajuster les paramtres
outlier_labels = lof.fit_predict(X)

# Supprimer les chantillons bruyants
X_cleaned = X[outlier_labels != -1]

# Crez un objet KMeans avec le nombre de clusters souhait
num_clusters = 2  # Un seul cluster
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Appliquez K-Means sur les donnes nettoyes
kmeans.fit(X_cleaned)

# Obtenez les labels des clusters pour chaque chantillon
labels = kmeans.labels_


# Rduction de dimension avec l'ACP (2 dimensions)
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_cleaned)  # Utilisez X_cleaned ici pour la dimension rduite

# Crez un DataFrame pour les donnes rduites en dimension
df_2d = pd.DataFrame(X_2d, columns=['Component 1', 'Component 2'])

# Ajoutez les labels de clusters au DataFrame
df_2d['Cluster'] = labels

# Crez un graphique de dispersion en colorant les points par cluster
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df_2d['Component 1'], df_2d['Component 2'], c=df_2d['Cluster'], cmap='viridis')
plt.title('Clustering avec K-Means (2D Projection)')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend(*scatter.legend_elements(), title='Clusters')
plt.show()
```

Le Silhouette Score est une mtrique qui mesure la qualit d'une segmentation (clustering) en valuant  quel point les objets au sein d'un mme cluster sont similaires les uns aux autres par rapport  la distance qui les spare des objets des autres clusters. Il fournit une mesure de la cohrence des clusters et peut aider  dterminer si les clusters sont bien dfinis et bien spars.

Le Silhouette Score est calcul pour chaque point de donnes dans l'ensemble de donnes et varie de -1  +1 :

Un score proche de +1 indique que le point est bien clusteris et est loin des points des autres clusters, ce qui signifie une segmentation de haute qualit.
Un score proche de 0 indique que le point est proche de la frontire entre deux clusters, ce qui signifie que la segmentation n'est pas trs claire.
Un score proche de -1 indique que le point a t attribu au mauvais cluster, ce qui signifie que la segmentation est incorrecte.

# Silhouette Score :

Le Silhouette Score mesure  quel point un point de donnes est similaire aux autres points dans son cluster par rapport  la distance moyenne entre ce point et les points du cluster voisin le plus proche. La formule est la suivante :

Silhouette Score pour un point i :

S(i) = (b(i) - a(i)) / max(a(i), b(i))

a(i) est la distance moyenne entre le point i et tous les autres points dans le mme cluster.
b(i) est la distance moyenne entre le point i et les points du cluster voisin le plus proche (c'est--dire le cluster diffrent de celui auquel appartient i).
Le Silhouette Score global est la moyenne des Silhouette Scores pour tous les points de donnes.

# l'indice de silhouette
est une mesure qui value la qualit globale d'un clustering en utilisant  la fois des informations intra-cluster et inter-cluster. C'est une mesure qui prend en compte  la fois la cohsion intra-cluster et la sparation inter-cluster.

Les composantes principales (component 1, component 2, et component 3) dans l'analyse en composantes principales (PCA) reprsentent des combinaisons linaires des variables originales de vos donnes qui capturent la plus grande part de la variance des donnes. Chaque composante principale est une nouvelle variable qui est une combinaison pondre des variables originales. 

```python
from sklearn.metrics import silhouette_score

# Calcul du coefficient de silhouette
silhouette_avg = silhouette_score(X_cleaned, labels)
print(f"Silhouette Score : {silhouette_avg}")
```
