```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des mod√®les
model_names = ["KNN", "SVM", "Logistic Regression", "Random Forest", "Deep Neural Network", "Neural Network", "BLSTM"]

# Nouvelles valeurs de m√©triques pour chaque mod√®le
accuracy = [0.951309, 0.789714, 0.811321, 0.987219, 0.970177, 0.943396, 0.962873]
precision = [0.948052, 0.730931, 0.826543, 0.981176, 0.971924, 0.952121, 0.972661]
recall = [0.957092, 0.930870, 0.797974, 0.994041, 0.969607, 0.936234, 0.954112]
f1_score = [0.952550, 0.818873, 0.812007, 0.987567, 0.970764, 0.944111, 0.963297]

# Largeur des barres
bar_width = 0.2

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Cr√©er la figure en sp√©cifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos pr√©f√©rences

# Cr√©er les barres pour chaque m√©trique
bar1 = plt.bar(index, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index + bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index + 2 * bar_width, recall, bar_width, label='Recall')
bar4 = plt.bar(index + 3 * bar_width, f1_score, bar_width, label='f1_score')

# D√©finir les √©tiquettes des mod√®les sur l'axe des x
plt.xlabel('Mod√®les')
plt.ylabel('Scores')
plt.title('Comparaison des mod√®les par m√©trique')
plt.xticks(index + 1.5 * bar_width, model_names)
plt.legend()

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(accuracy):
    plt.text(index[i] - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(precision):
    plt.text(index[i] + bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(recall):
    plt.text(index[i] + 2 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(f1_score):
    plt.text(index[i] + 3 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Donn√©es du tableau
data = {
    "Model": ["KNN", "SVM", "Logistic Regression", "Random forest", "Deep neural network", "Convolutional neural network", "BLSTM"],
    "Accuracy": [0.951309, 0.789714, 0.811321, 0.987219, 0.964394, 0.961351, 0.934571],
    "Precision": [0.948052, 0.730931, 0.826543, 0.981176, 0.954043, 0.938879, 0.951822],
    "Recall": [0.957092, 0.930870, 0.797974, 0.994041, 0.977354, 0.988677, 0.918355],
    "F1 Score": [0.952550, 0.818873, 0.812007, 0.987567, 0.965558, 0.963135, 0.934789],
    "AUC": [0.951183, 0.786642, 0.811611, 0.987070, 0.964112, 0.960756, 0.934924]
}

# Cr√©er un DataFrame √† partir des donn√©es
df = pd.DataFrame(data)

# Liste des noms des mod√®les
model_names = df["Model"]

# M√©triques √† afficher
metrics = ["Accuracy", "Precision", "Recall", "F1 Score", "AUC"]

# Largeur des barres
bar_width = 0.15

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Cr√©er la figure en sp√©cifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos pr√©f√©rences

# Cr√©er les barres pour chaque m√©trique
for i, metric in enumerate(metrics):
    values = df[metric]
    plt.bar(index + i * bar_width, values, bar_width, label=metric)

# D√©finir les √©tiquettes des mod√®les sur l'axe des x avec rotation
plt.xlabel('Mod√®les')
plt.ylabel('Scores')
plt.title('Comparaison des mod√®les par m√©trique')
plt.xticks(index + bar_width * (len(metrics) - 1) / 2, model_names, rotation=45)  # Rotation personnalis√©e ici
plt.legend()

# Ajouter les valeurs au-dessus des barres avec une taille de police r√©duite
for i, metric in enumerate(metrics):
    values = df[metric]
    for j, v in enumerate(values):
        plt.text(index[j] + i * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')

# Afficher le tableau
print(df)

# Afficher le graphique
plt.tight_layout()
plt.show()
```
