```python
import pandas as pd

# Importez les trois datasets depuis leurs fichiers CSV respectifs
data1 = pd.read_csv('knn_results.csv')
data2 = pd.read_csv('svm_results.csv')
data3 = pd.read_csv('lr_results.csv')
data4 = pd.read_csv('randfor_results.csv')
data5 = pd.read_csv('destree_results.csv')
data6 = pd.read_csv('xgb_results.csv')
data7 = pd.read_csv('dnn_results.csv')
data8 = pd.read_csv('cnn_results.csv')
data9 = pd.read_csv('blstm_results.csv')


# Concaténez les trois DataFrames en un seul
all_data = pd.concat([data1, data2, data3, data4, data5, data6, data7 , data8 , data9 ], ignore_index=True)

# Affichez le DataFrame résultant
all_data.head(n=10)

```

```python
# Enregistrez le DataFrame résultant dans un fichier CSV
all_data.to_csv('evaluation_final.csv', index=False)  # index=False pour exclure l'index dans le fichier CSV
```
