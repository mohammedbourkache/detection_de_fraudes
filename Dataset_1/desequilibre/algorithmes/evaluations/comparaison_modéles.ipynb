```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modles
model_names = [
    "KNN",
    "SVM",
    "Logistic Regression",
    "Random Forest",
    "Deep Neural Network",
    "Convolutional Neural Network",
    "BLSTM",
]

# Exemple de valeurs de mtriques pour chaque modle
accuracy = [0.992468, 0.983540, 0.977158, 0.995152, 0.991839, 0.992268, 0.979229]
precision = [0.880674, 0.990066, 0.814286, 0.948769, 0.865248, 0.929682, 0.619707]
recall = [0.815616, 0.359159, 0.136937, 0.856456, 0.806006, 0.754354, 0.483483]
f1_score = [0.846897, 0.527104, 0.234447, 0.900253, 0.834577, 0.832891, 0.543185]
auc = [0.906359, 0.679532, 0.568059, 0.927622, 0.901358, 0.876429, 0.737853]

# Largeur des barres
bar_width = 0.15

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Crer la figure en spcifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos prfrences

# Crer les barres pour chaque mtrique
bar1 = plt.bar(index - 2 * bar_width, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index - bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index, recall, bar_width, label='Recall')
bar4 = plt.bar(index + bar_width, f1_score, bar_width, label='F1 Score')
bar5 = plt.bar(index + 2 * bar_width, auc, bar_width, label='AUC')

# Faire pivoter les tiquettes des modles
plt.xticks(index, model_names, rotation=45, fontsize=10, ha="right")

# Ajouter les valeurs au-dessus des barres avec une police plus petite
def add_values_to_bars(bar_list):
    for bars in bar_list:
        for bar in bars:
            height = bar.get_height()
            plt.annotate(
                f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center',
                fontsize=8,  # Taille de la police rduite
            )

add_values_to_bars([bar1, bar2, bar3, bar4, bar5])

# Ajouter un titre
plt.title('Comparaison des modles par mtrique')

# Ajouter des tiquettes d'axe
plt.xlabel('Modles')
plt.ylabel('Scores')

# Afficher la lgende
plt.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```
