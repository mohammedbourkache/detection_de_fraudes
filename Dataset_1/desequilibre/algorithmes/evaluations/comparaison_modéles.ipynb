```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modles
model_names = ["KNN", "SVM", "Logistic Regression", "Random Forest", "Deep Neural Network", "CNN", "BLSTM"]

# Exemple de valeurs de mtriques pour chaque modle
accuracy = [0.992468, 0.983540, 0.762269, 0.995152, 0.974473, 0.974458, 0.979198]
precision = [0.880674, 0.990066, 0.000289, 0.948769, 1.000000, 0.000000, 0.619490]
recall = [0.815616, 0.359159, 0.002402, 0.856456, 0.000601, 0.000000, 0.481081]
f1_score = [0.846897, 0.527104, 0.000516, 0.900253, 0.001200, 0.000000, 0.541582]

# Largeur des barres
bar_width = 0.2

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Crer la figure en spcifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos prfrences

# Crer les barres pour chaque mtrique
bar1 = plt.bar(index, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index + bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index + 2 * bar_width, recall, bar_width, label='Recall')
bar4 = plt.bar(index + 3 * bar_width, f1_score, bar_width, label='f1_score')

# Dfinir les tiquettes des modles sur l'axe des x
plt.xlabel('Modles')
plt.ylabel('Scores')
plt.title('Comparaison des modles par mtrique')
plt.xticks(index + 1.5 * bar_width, model_names)
plt.legend()

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(accuracy):
    plt.text(index[i] - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(precision):
    plt.text(index[i] + bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(recall):
    plt.text(index[i] + 2 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(f1_score):
    plt.text(index[i] + 3 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')

# Afficher le graphique
plt.tight_layout()
plt.show()
```
