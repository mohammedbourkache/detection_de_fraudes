```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modèles
model_names = ["KNN", "SVM", "Logistic Regression", "Random Forest", "Deep Neural Network", "Neural Network", "BLSTM"]

# Exemple de valeurs de métriques pour chaque modèle (à remplacer par vos propres valeurs)
accuracy = [0.996054, 0.789714, 0.987719, 0.997141, 0.987734, 0.987726, 0.987788]
precision = [0.870255, 0.730931, 0.000000, 0.939811, 1.000000, 0.666667, 0.909091]
recall = [0.797596, 0.930870, 0.000000, 0.819734, 0.001265, 0.001265, 0.006325]
f1_score = [0.832343, 0.818873, 0.000000, 0.875676, 0.002527, 0.002525, 0.012563]

# Largeur des barres
bar_width = 0.2

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Créer la figure en spécifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos préférences

# Créer les barres pour chaque métrique
bar1 = plt.bar(index, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index + bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index + 2 * bar_width, recall, bar_width, label='Recall')
bar4 = plt.bar(index + 3 * bar_width, f1_score, bar_width, label='f1_score')

# Définir les étiquettes des modèles sur l'axe des x
plt.xlabel('Modèles')
plt.ylabel('Scores')
plt.title('Comparaison des modèles par métrique')
plt.xticks(index + 1.5 * bar_width, model_names)
plt.legend()

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(accuracy):
    plt.text(index[i] - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(precision):
    plt.text(index[i] + bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(recall):
    plt.text(index[i] + 2 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(f1_score):
    plt.text(index[i] + 3 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')

# Afficher le graphique
plt.tight_layout()
plt.show()
```
