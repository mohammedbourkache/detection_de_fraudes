```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modèles
model_names = ["Model 1", "Model 2", "Model 3", "Model 4", "Model 5", "Model 6", "Model 7", "Model 8"]

# Exemple de valeurs de métriques pour chaque modèle (à remplacer par vos propres valeurs)
accuracy = np.random.rand(8)  # Remplacez par vos valeurs d'accuracy
precision = np.random.rand(8)  # Remplacez par vos valeurs de précision
recall = np.random.rand(8)  # Remplacez par vos valeurs de rappel
roc_auc = np.random.rand(8)  # Remplacez par vos valeurs de ROC AUC

# Largeur des barres
bar_width = 0.2

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Créer la figure et les sous-graphiques
fig, ax = plt.subplots()

# Créer les barres pour chaque métrique
bar1 = ax.bar(index, accuracy, bar_width, label='Accuracy')
bar2 = ax.bar(index + bar_width, precision, bar_width, label='Precision')
bar3 = ax.bar(index + 2 * bar_width, recall, bar_width, label='Recall')
bar4 = ax.bar(index + 3 * bar_width, roc_auc, bar_width, label='ROC AUC')

# Définir les étiquettes des modèles sur l'axe des x
ax.set_xlabel('Modèles')
ax.set_ylabel('Scores')
ax.set_title('Comparaison des modèles par métrique')
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(model_names)
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modèles
model_names = ["Model 1", "Model 2", "Model 3", "Model 4", "Model 5", "Model 6", "Model 7", "Model 8"]

# Exemple de valeurs de métriques pour chaque modèle (à remplacer par vos propres valeurs)
accuracy = np.random.rand(8)  # Remplacez par vos valeurs d'accuracy
precision = np.random.rand(8)  # Remplacez par vos valeurs de précision
recall = np.random.rand(8)  # Remplacez par vos valeurs de rappel
roc_auc = np.random.rand(8)  # Remplacez par vos valeurs de ROC AUC

# Largeur des barres
bar_width = 0.2

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Créer la figure en spécifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos préférences

# Créer les barres pour chaque métrique
bar1 = plt.bar(index, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index + bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index + 2 * bar_width, recall, bar_width, label='Recall')
bar4 = plt.bar(index + 3 * bar_width, roc_auc, bar_width, label='ROC AUC')

# Définir les étiquettes des modèles sur l'axe des x
plt.xlabel('Modèles')
plt.ylabel('Scores')
plt.title('Comparaison des modèles par métrique')
plt.xticks(index + 1.5 * bar_width, model_names)
plt.legend()

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(accuracy):
    plt.text(index[i] - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(precision):
    plt.text(index[i] + bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(recall):
    plt.text(index[i] + 2 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')
for i, v in enumerate(roc_auc):
    plt.text(index[i] + 3 * bar_width - 0.05, v + 0.01, f'{v:.2f}', fontsize=9, ha='center', va='bottom')

# Afficher le graphique
plt.tight_layout()
plt.show()
```
