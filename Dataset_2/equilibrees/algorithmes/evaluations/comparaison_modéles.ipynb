```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des modèles
model_names = ["KNN", "SVM", "Logistic regression", "Random forest", "Deep neurual network", "Convolutional neurual network", "BLSTM"]

# Exemple de valeurs de métriques pour chaque modèle (à remplacer par vos propres valeurs)
accuracy = [0.894066, 0.618067, 0.661661, 0.980258, 0.767547, 0.500008, 0.886482]
precision = [0.843407, 0.636276, 0.678199, 0.979511, 0.707717, 0.500008, 0.905616]
recall = [0.967830, 0.551287, 0.615280, 0.981037, 0.911578, 1.000000, 0.862899]
roc_auc = [0.894065, 0.618068, 0.661662, 0.980258, 0.767544, 0.500000, 0.886482]
f1_score = [0.901345, 0.590740, 0.645209, 0.980273, 0.796815, 0.666674, 0.883742]

# Largeur des barres
bar_width = 0.15

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Créer la figure en spécifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos préférences

# Créer les barres pour chaque métrique
bar1 = plt.bar(index - 1.5 * bar_width, accuracy, bar_width, label='Accuracy')
bar2 = plt.bar(index - 0.5 * bar_width, precision, bar_width, label='Precision')
bar3 = plt.bar(index + 0.5 * bar_width, recall, bar_width, label='Recall')
bar4 = plt.bar(index + 1.5 * bar_width, roc_auc, bar_width, label='ROC AUC')
bar5 = plt.bar(index + 2.5 * bar_width, f1_score, bar_width, label='F1 Score')

# Définir les étiquettes des modèles sur l'axe des x avec rotation
plt.xlabel('Modèles')
plt.ylabel('Scores')
plt.title('Comparaison des modèles par métrique')
plt.xticks(index, model_names, rotation=45)  # Rotation personnalisée ici
plt.legend()

# Ajouter les valeurs au-dessus des barres avec une taille de police réduite
for i, v in enumerate(accuracy):
    plt.text(index[i] - 1.5 * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')
for i, v in enumerate(precision):
    plt.text(index[i] - 0.5 * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')
for i, v in enumerate(recall):
    plt.text(index[i] + 0.5 * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')
for i, v in enumerate(roc_auc):
    plt.text(index[i] + 1.5 * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')
for i, v in enumerate(f1_score):
    plt.text(index[i] + 2.5 * bar_width, v + 0.01, f'{v:.2f}', fontsize=7, ha='center', va='bottom')

# Afficher le graphique
plt.tight_layout()
plt.show()
```
