```python
import pandas as pd

data = pd.read_csv('evaluation_final.csv')
```

```python
df = data.copy()
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Liste des noms des mod√®les
model_names = df["Model"]

# M√©triques √† afficher
metrics = ["Accuracy", "Precision", "Recall", "F1 Score", "AUC"]

# Largeur des barres
bar_width = 0.15

# Position des barres sur l'axe des x
index = np.arange(len(model_names))

# Cr√©er la figure en sp√©cifiant la taille
plt.figure(figsize=(12, 6))  # Ajustez la taille selon vos pr√©f√©rences

# Cr√©er les barres pour chaque m√©trique
for i, metric in enumerate(metrics):
    values = df[metric]
    plt.bar(index + i * bar_width, values, bar_width, label=metric)

# D√©finir les √©tiquettes des mod√®les sur l'axe des x avec rotation
plt.xlabel('Mod√®les')
plt.ylabel('Scores')
plt.title('Comparaison des mod√®les par m√©trique')
plt.xticks(index + bar_width * (len(metrics) - 1) / 2, model_names, rotation=45)  # Rotation personnalis√©e ici

# Ajouter les valeurs au-dessus des barres avec une taille de police r√©duite et des valeurs arrondies
for i, metric in enumerate(metrics):
    values = df[metric]
    for j, v in enumerate(values):
        plt.text(index[j] + i * bar_width, v + 0.01, f'{v:.2f}', fontsize=5.5, ha='center', va='bottom')

# Afficher la l√©gende avec une taille de police r√©duite
plt.legend(fontsize=6)

# Enregistrer le graphique au format PDF
plt.tight_layout()
plt.savefig('comparaison_ds2.pdf', format='pdf')

# Afficher le graphique
plt.show()
```
